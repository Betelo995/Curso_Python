#### Máquina de vueltos en autobuses


Para el ejemplo de las funciones vamos a realizar una calculadora de vueltos para una empresa de autobuses, que (en un futuro utópico) instaló un sistema de pagos automático y moderno dentro del sistema de buses. La calculadora de vueltos funcionará dentro de una máquina especial, la cual debe recibir la plata y decirle al pasajero cual será su vuelto en monedas de $500$, $100$, $50$, $25$, $10$ y $5$, además de proceder a darle el vuelto. 

En este caso es únicamente necesario que cuando el código se ejecute despliegue la cantidad de cada moneda o en caso de que el dinero esté completo o incompleto lo indique. 


Primero debemos comenzar por invocar la función como tal y darle un nombre, además recordar que la función debe indicar cuantos parámetros recibe:

Los casos más sencillos, donde el vuelto es igual a cero o menor que cero, es donde únicamente debemos decirle al usuario del bus, que o no hay vuelto ó que necesita dar más dinero, ya que el pasaje se encuentra incompleto. Para el caso de que el vuelto sea mayor que cero:

Ahora en el caso de que el vuelto sea mayor que cero, tenemos que obtener la cantidad de cada una de las monedas que la máquina va a devolver al usuario. Para ello debemos hacer un algoritmo que nos de un valor exacto y que vaya reduciendo hasta que se complete todo el valor del vuelto. La forma más sencilla de lograr esto es comenzar con la moneda más grande, que sería la de $500$ y a partir de ello, ir hacia abajo obteniendo cantidades. Las herramientas que nos van a dar la cantidad de monedas de cada una son la división entera `a//b` y el módulo `a%b`, de las cuales de la división obtenemos la cantidad de $b$ que hay en $a$ y del módulo, obtenemos el residuo de la división entre $a$ y $b$. Con esto mismo, obtenemos de la primera la cantidad de monedas que podemos sacar de la denominación actual y  de la segunda lo que sobra cuando quitamos esto. para un caso se vería así: 

```Python
def maquina_pasajes(pasaje, pago):
    vuelto = pago - pasaje 
    if vuelto == 0:
        print("Pagó de forma completa, por lo tanto no hay vuelto")
       
    elif vuelto > 0:
        print("Su vuelto es: %d" %(vuelto))
        print("Las monedas de su vuelto son: ")
        monedas_500 = vuelto//500 #Obtenemos la cantidad de monedas de 500 que debemos darle al usuario

```

Ahora únicamente debemos crear un programa principal que reciba la cantidad que cuesta el pasaje, indicado por el chofer y además cuando recibe de plata la máquina por parte del usuario, obteniendo así el siguiente programa: 


````python
def maquina_pasajes(pasaje, pago):
    vuelto = pago - pasaje
    if vuelto == 0: 
        print("Pagó de forma completa, por lo tanto no hay vuelto")
    elif vuelto > 0: 
        print("Su vuelto es: %d" %(vuelto))
        print("Las monedas de su vuelto son: ")
        monedas_500 = vuelto//500
        residuo = vuelto%500
        print("500: %d" %(monedas_500))
        monedas_100 = residuo//100
        print("100: %d" %(monedas_100))
        residuo = residuo%100
        monedas_50 = residuo//50
        print("50: %d" %(monedas_50))
        residuo = residuo%50
        monedas_25 = residuo//25
        print("25: %d" %(monedas_25))
        residuo = residuo%25
        monedas_10 = residuo//10
        print("10: %d" %(monedas_10))
        residuo = residuo%10
        monedas_5 = residuo//5
        print("5: %d" %(monedas_5))
    
    elif vuelto < 0:
        print("La cantidad recibida es insuficiente, ingrese más dinero por favor")

try: 
    iniciar = float(input("Ingrese el costo del pasaje: "))
    ingreso = float(input("Monto con el que pagó el pasajero: "))
    maquina_pasajes(iniciar, ingreso)
except:
    print("Sus datos son inválidos, intente de nuevo.")
````

Con esto podemos ver la facilidad que trae una función a un código, ya que por ejemplo digamos que la red de autobuseros completa está conectada a un servidor donde el servidor mismo corre el programa y así se ejecuta en cada bus, en vez de implementar el bloque de código completo de la función, cada máquina simplemente envía al servidor una petición de correr `maquina_pasajes(valor, pago)` y así, se puede lograr un código más eficiente, corto y optimizado.
